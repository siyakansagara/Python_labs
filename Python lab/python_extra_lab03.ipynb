### Q1:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





```python
h = int(input("Enter head"))
l = int(input("Enter legs"))
d = (l-2*h)/2
hen = h-d
print("dogs",d ,"chicken",hen)
```

### Q2:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

```python
n1 = int(input("Enter numerator of first number: "))
d1 = int(input("Enter denominator of first number: "))
n2 = int(input("Enter numerator of second number: "))
d2 = int(input("Enter denominator of second number: "))

sum_numerator = (n1 * d2) + (n2 * d1)
sum_denominator = d1 * d2
print("Sum of fractions =", sum_numerator, "/", sum_denominator)
```

### Q3:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.
Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

```python
import math
H = int(input("Enter height of tank (cm): "))
L = int(input("Enter length of tank (cm): "))
B = int(input("Enter breadth of tank (cm): "))

h = int(input("Enter height of glass (cm): "))
r = int(input("Enter radius of glass (cm): "))

tank_volume = H * L * B

glass_volume = math.pi * r * r * h

no_of_glasses = tank_volume // glass_volume

print("Volume of tank =", tank_volume, "cm続")
print("Volume of one glass =", round(glass_volume, 2), "cm続")
print("Number of glasses of milk =", int(no_of_glasses))
```
