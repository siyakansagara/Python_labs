### 01) WAP to find sum of all the elements in a List.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(int(input("Enter element: ")))
print("Sum of elements:", sum(lst))
```

### 02) WAP to find largest element in a List.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(int(input("Enter element: ")))
largest = lst[0]
for i in lst:
    if i > largest:
        largest = i
print("Largest element:", largest)
```

### 04) WAP to interchange first and last elements in a list.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(input("Enter element: "))
print("Original List:", lst)
if n > 1:
    temp = lst[0]
    lst[0] = lst[n - 1]
    lst[n - 1] = temp
print("List after interchange:", lst)
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    element = int(input(f"Enter element {i+1}: "))
    lst.append(element)
k = int(input("Enter split position: "))
result = []
for i in range(k, n):
    result.append(lst[i])
print("Modified List:", result)
```

### 06) WAP to interchange the elements on two positions entered by a user.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    element = int(input(f"Enter element {i+1}: "))
    lst.append(element)
pos1 = int(input("Enter first position to swap: "))
pos2 = int(input("Enter second position to swap: "))
temp = lst[pos1]
lst[pos1] = lst[pos2]
lst[pos2] = temp
print("List after swapping:", lst)
```

### 07) WAP to reverse the list entered by user.

### 08) WAP to print even numbers in a list.

### 11) WAP to print all odd numbers in a given range.

### 12) WAP to count occurrences of an element in a list.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(input("Enter element: "))
element = input("Enter element to count: ")
count = 0
for i in lst:
    if i == element:
        count += 1
print("Occurrences of", element, ":", count)
```

### 13) WAP to find second largest number in a list.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(int(input("Enter element: ")))
largest = second_largest = -999999
for i in lst:
    if i > largest:
        second_largest = largest
        largest = i
    elif i > second_largest and i != largest:
        second_largest = i
print("Second largest number:", second_largest)
```

### 14) WAP to extract elements with frequency greater than K.

```python
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(input("Enter element: "))
k = int(input("Enter value of K: "))
result = []
for i in lst:
    if lst.count(i) > k and i not in result:
        result.append(i)
print("Elements with frequency greater than K:", result)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
n = int(input("Enter number of fruits: "))
fruits = []
for i in range(n):
    fruits.append(input("Enter fruit name: "))
b_fruits = [fruit for fruit in fruits if fruit.lower().startswith('b')]
print("Fruits starting with 'b':", b_fruits)
```

### 17) WAP to create a list of common elements from given two lists.

```python
n1 = int(input("Enter number of elements in list 1: "))
list1 = []
for i in range(n1):
    list1.append(input("Enter element: "))
n2 = int(input("Enter number of elements in list 2: "))
list2 = []
for i in range(n2):
    list2.append(input("Enter element: "))
common = []
for i in list1:
    if i in list2 and i not in common:
        common.append(i)
print("Common elements:", common)
```
