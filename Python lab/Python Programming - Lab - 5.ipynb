### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = input("Enter the string: ")

uppercase = 0
lowercase = 0
digits = 0
special = 0

for ch in s:
    if 'A' <= ch <= 'Z':
        uppercase += 1
    elif 'a' <= ch <= 'z':
        lowercase += 1
    elif '0' <= ch <= '9':
        digits += 1
    else:
        special += 1

print("Uppercase:", uppercase ,"Lowercase:", lowercase ,"Digits:", digits ,"Special:", special)
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
reversed_words = []
for word in words:
    reversed_words.append(word[::-1])

result = " ".join(reversed_words)
print("Output:", result)
```

### 3) Case Pattern Identifier
- `PYTHON` → Uppercase
- `python` → Lowercase
- `Python Programming` → Title Case

```python
inputs = []

for i in range(1, 4):
    text = input(f"Enter string {i}: ")
    inputs.append(text)

for text in inputs:
    if text.isupper():
        print(f"{text} → Uppercase")
    elif text.islower():
        print(f"{text} → Lowercase")
    elif text.istitle():
        print(f"{text} → Title Case")
    else:
        print(f"{text} → Mixed Case")
```

### 8) Replace Vowels
- `Programming` → `Pr*gr*mm*ng`
- `Education` → `*d*c*t**n`
- `Sky` → `Sky`

```python
s = input("Enter string: ")
vowels = "aeiouAEIOU"
result = ""
for ch in s:
    if ch in vowels:
        result += "*"
    else:
        result += ch
print(result)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter string: ")
result = ""
count = 1
for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        result += s[i-1] + str(count)
        count = 1
result += s[-1] + str(count)
print(result)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter string: ")
result = ""
for ch in s:
    if ch.isupper():
        result += ch.lower()
    elif ch.islower():
        result += ch.upper()
    else:
        result += ch

print(result)
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit
3. length ≥ 6

- `User123` → Valid
- `123User` → Invalid
- `Us@12` → Invalid

```python
username = input("Enter username: ")

if (username.isalnum() and
    not username[0].isdigit() and
    len(username) >= 6):
    print("Valid")
else:
    print("Invalid")
```

### 13) Longest Word Finder
- `Python programming is interesting` → programming
