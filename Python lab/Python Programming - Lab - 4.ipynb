### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input("Enter number:"))
sum = 0

for i in range(1, n+1):
    if i % 2 == 0:
        sum -= i
    else:
        sum += i

print("Sum =", sum)
```

### 10) WAP to print GCD of given two numbers.

```python
a = int(input("Enter a number:"))
b = int(input("Enter a number:"))
while a != b:
    if a > b:
        a -= b
    else:
        b -= a

print("GCD =", a)
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter number:"))
flag = True

for i in range(2, n):
    if n % i == 0:
        flag = False

if flag:
    print("Prime")
else:
    print("Not Prime")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter number:"))
sum = 0
while n > 0:
    sum += n % 10
    n //= 10

print("Sum of digits =", sum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter number:"))
temp = n
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num = int(input("Enter a number: "))
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10
    
if sum == num:
    print("Armstrong Number")
else:
    print("not Armstrong Number")
```
